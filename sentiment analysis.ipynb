```python
import pandas as pd
```

```python
data = pd.read_csv('Dataset .csv')
```

```python
data.head()
```

```python
restaurant_count = data.groupby('City').size().reset_index(name="Restaurant_count")
```

```python
restaurant_count
```

```python
#city with most restaurants
city_with_most_restaurants = restaurant_count.loc[restaurant_count["Restaurant_count"].idxmax()]
```

```python
city_with_most_restaurants
```

```python
average_rating = data.groupby("City")["Aggregate rating"].mean().reset_index(name="Average_rating")
```

```python
city_with_highest_average_rating = average_rating.loc[average_rating["Average_rating"].idxmax()]
```
